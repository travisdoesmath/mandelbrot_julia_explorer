```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
class Poly():
    
    def __init__(self, terms):
        self.terms = {}
        for k, v in terms.items():
            self.terms[k] = self.terms.get(k, 0) + v
        self.terms = {x:y for x,y in self.terms.items() if y != 0}
            
            
    def f(self, x, y):
        v = 0
        for (x_, y_), k in self.terms.items():
            v += k*x**x_*y**y_
        return v
    
    def dy(self):
        terms = {}
        for (x, y), v in self.terms.items():
            terms[(x, y-1)] = y * v
        return Poly(terms)
    
    def dx(self):
        terms = {}
        for (x, y), v in self.terms.items():
            terms[(x-1, y)] = x * v
        return Poly(terms)    
    
    def dydx(self):
        return RationalFunction(-self.dx(), self.dy())
            
    def __add__(self, o):
        terms = {}
        for k, v in self.terms.items():
            terms[k] = terms.get(k, 0) + v
        for k, v in o.terms.items():
            terms[k] = terms.get(k, 0) + v
        return Poly(terms)
    
    def __mul__(self, o):
        terms = {}
        for (x1, y1), v1 in self.terms.items():
            for (x2, y2), v2 in o.terms.items():
                terms[(x1+x2, y1+y2)] = terms.get((x1+x2, y1+y2), 0) + v1 * v2
        return Poly(terms)
    
    def __neg__(self):
        terms = {}
        for k, v in self.terms.items():
            print(k, v)
            terms[k] = -v
        return Poly(terms)
    
    def __sub__(self, o):
        return self + -o
    
    def __repr__(self):
        s = ''
        for k, v in self.terms.items():
            if v != 0:
                term = ''
                if k[0] != 0: 
                    term += 'x'
                    if k[0] != 1: term += f'^{k[0]}' 
                if k[1] != 0: 
                    term += 'y'
                    if k[1] != 1: term += f'^{k[1]}'
                s = f'{v if v != 1 else ""}{term}' if s == '' else ' + '.join([s, f'{v if v != 1 else ""}{term}'])
        return s
    
class RationalFunction():
    
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator
        
    def f(self, x, y):
        if self.denominator.f(x, y) == 0:
            if self.numerator.f(x, y) > 0:
                return np.inf
            elif self.numerator.f(x, y) < 0:
                return -np.inf
            else:
                raise ValueError("Zero divided by Zero encountered")

        else:
            return self.numerator.f(x, y) / self.denominator.f(x, y)
        
    def __repr__(self):
        return f'{self.numerator} / {self.denominator}'

class Complex():
    
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
        
    def magnitude_sq(self):
        return self.real * self.real + self.imag * self.imag
    
    def __add__(self, o):
        return Complex(self.real + o.real, self.imag + o.imag)
    
    def __mul__(self, o):
        real = self.real*o.real - self.imag*o.imag
        imag = self.real*o.imag + self.imag*o.real
        return Complex(real, imag)
    
    def __repr__(self):
        return f'{self.real} + i({self.imag})'
```

```python
def cpoint(poly, x, y, epsilon=0.0000001):
    F = poly.f
    F_x = poly.dx().f
    F_y = poly.dy().f
    iter = 0
    cutoff = 10
    d = np.inf
    while (d > epsilon and iter < cutoff):
        new_x = x - (F(x, y)) / (F_x(x, y)**2 + F_y(x, y)**2) * F_x(x, y)
        new_y = y - (F(x, y)) / (F_x(x, y)**2 + F_y(x, y)**2) * F_y(x, y)
        d = (x - new_x)**2 + (y - new_y)**2
        x = new_x
        y = new_y
    return (x, y)
```
