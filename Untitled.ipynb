```python
import pandas as pd
from matplotlib import pyplot as plt
import math
import numpy as np
```

```python
class Complex():
    
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag
                
    def r(self):
        return (self.real * self.real + self.imag * self.imag)**(1/2)
    
    def theta(self):
        return math.atan2(self.imag, self.real)
    
    def sqrt(self):
        return [Complex( self.r()**(1/2)*math.cos(self.theta()/2),
                         self.r()**(1/2)*math.sin(self.theta()/2)),
                Complex(-self.r()**(1/2)*math.cos(self.theta()/2),
                        -self.r()**(1/2)*math.sin(self.theta()/2))]
        
    def __add__(self, o):
        return Complex(self.real + o.real, self.imag + o.imag)
    
    def __neg__(self):
        return Complex(-self.real, -self.imag)
    
    def __sub__(self, o):
        return self + -o
    
    def __mul__(self, o):
        real = self.real*o.real - self.imag*o.imag
        imag = self.real*o.imag + self.imag*o.real
        return Complex(real, imag)
    
    def __repr__(self):
        return f'{self.real} + {self.imag}i'
```

```python
def f(z):
    return z * z + z
    
def f_inverse(z):
    d = Complex(1, 0) + Complex(4, 0) * z
    d_sqrts = d.sqrt()
    return [ d_sqrts[0] * Complex(0.5, 0) - Complex(0.5, 0),
             d_sqrts[1] * Complex(0.5, 0) - Complex(0.5, 0)]
            
```

```python
n = 128
zs = []
for angle in np.arange(0, 2 * np.pi, np.pi/n):
    zs.append(Complex(2 * math.cos(angle), 2 * math.sin(angle)))
zs
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]
```

```python
zs = flatten([f_inverse(z) for z in zs])

xs = [z.real for z in zs]
ys = [z.imag for z in zs]
plt.scatter(xs, ys)
```
